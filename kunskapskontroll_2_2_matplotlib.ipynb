### Explain shortly what a figure, axes, axis and an artist is in Matplotlib.

**Figure** — is like a canvas an include everything

**Axes** — is a _single_ plot/graph within the figure (including lables and titles etc)

**Axis** — is the actual axis for within the _axes_. Axis handles scaling

**Artist** — the base class for all elements that are visible in the figure

### Create a plot of the function y = x^2 [from -4 to 4, hint use the np.linspace function] both in the object-oriented approach and the pyplot approach. Your plot should have a title and axis-labels.

```python
x = np.linspace(-4,4,100)
y = x**2

## "object oriented approach"
fig, ax = plt.subplots()
ax.plot(x, y)
ax.set_title("X Squared using ax")
ax.set_xlabel("x")
ax.set_ylabel("f(x) = x^2")
ax.grid(True)
ax.tick_params(axis='both', which='both')
```

```python
## "pyplot approach"
plt.plot(x, y)
plt.title("X Squared using pyplot approach")
plt.xlabel("x")
plt.ylabel("y = x^2")
plt.grid(True)
plt.tick_params(axis='both', which='both')
```

### Create a figure containing 2  subplots where the first is a scatter plot and the second is a bar plot. You have the data below. 

```python
# Data for scatter plot
np.random.seed(15)
random_data_x = np.random.randn(1000)
random_data_y = np.random.randn(1000)
x = np.linspace(-2, 2, 100)
y = x**2

# Data for bar plot
fruit_data = {'grapes': 22, 'apple': 8, 'orange': 15, 'lemon': 20, 'lime': 25}
names = list(fruit_data.keys())
values = list(fruit_data.values())
```

```python
fig, ax = plt.subplots(1, 2, figsize=(10,5))
fig.subplots_adjust(wspace=0.3)

## scatter
ax[0].scatter(random_data_x, random_data_y)

## bar diagram
ax[1].bar(names, values)
ax[1].set_xlabel("Fruits")
ax[1].set_ylabel("Quantity")
```
